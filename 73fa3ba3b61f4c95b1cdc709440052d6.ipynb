```python
##Create a function to calculate distance between two coordinates using the haversine formula
##Input parameters are two coordinates 
import math as m

##Defining haver as x
def haver(x):
    return (m.sin(x/2))**2

##Assigning values to lists A and B
def calcDist(A,B):
    Radius = 6371
    a = haver((B[0]-A[0])) + (m.cos(A[0])*m.cos(B[0])*haver(B[1]-A[1]))
    c = 2 * m.atan2(m.sqrt(a),m.sqrt(1-a))
    d = Radius * c
    return d

#Theoretical values:
#A = [23,34]
#B = [16,23]
#print(calcDist(A,B))
```
